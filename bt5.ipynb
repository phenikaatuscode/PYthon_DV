```python
import pandas as pd
import matplotlib.pyplot as plt

url = 'https://raw.githubusercontent.com/thieu1995/csv-files/main/data/DV-course/toyota.csv'  
df = pd.read_csv(url)

df.dropna

df.head()
```

### -------------------- Bài Tập 1 ------------------------ ###

```python
#1.Price và Mileage
plt.figure(figsize=(10, 6))  
plt.scatter(df['price'], df['mileage'], c='blue', edgecolors='k', alpha=0.7, s=50)  

plt.title('Scatter plot of Price vs Mileage', fontsize=16, fontweight='bold')  
plt.xlabel('Price (in USD)', fontsize=14)  
plt.ylabel('Mileage (in miles)', fontsize=14)  
plt.grid(True, which='both', linestyle='--', linewidth=0.5)  

plt.show()

```

***Nhận xét biểu đồ #1***:

Biểu đồ **Price vs Mileage** phản ánh mối quan hệ nghịch biến rõ ràng giữa giá cả (Price) và số dặm đã đi (Mileage). Cụ thể:

**Mối quan hệ nghịch đảo**: Khi số dặm đi tăng lên, giá của xe giảm đáng kể. Điều này thể hiện rằng những chiếc xe đã qua sử dụng nhiều thường có giá trị giảm đi. Xe đã chạy nhiều (mileage cao) thường bị coi là cũ hơn và có nguy cơ hỏng hóc cao hơn, dẫn đến việc người mua sẵn sàng trả ít tiền hơn.
  
**Tốc độ giảm giá mạnh**: Đặc biệt ở những khoảng mileage thấp đến trung bình (dưới 100,000 miles), có thể thấy giá xe giảm rất nhanh. Điều này có thể lý giải rằng các xe mới, hoặc xe chạy ít, giữ giá tốt hơn rất nhiều so với xe đã qua sử dụng lâu dài. 

**Biến động lớn ở giá cao**: Trong biểu đồ cũng có một số điểm ngoại lệ, ví dụ các xe có giá cao dù mileage lớn. Đây có thể là những mẫu xe cao cấp hoặc xe đã được bảo dưỡng đặc biệt, cho phép chúng giữ giá tốt hơn bình thường.

Kết luận: **Biểu đồ nhấn mạnh tầm quan trọng của mileage trong việc định giá xe**, với mileage là một yếu tố chính trong việc dự đoán giá trị của xe trên thị trường. Xe mới hơn, ít chạy hơn có xu hướng giữ giá cao hơn đáng kể so với những xe đã qua sử dụng nhiều.

```python
#2 Price vs Tax
plt.figure(figsize=(10, 6))  

plt.scatter(df['price'], df['tax'], c='green', edgecolors='k', alpha=0.7, s=70)  

plt.title('Scatter plot of Price vs Tax', fontsize=16, fontweight='bold')  
plt.xlabel('Price (in USD)', fontsize=14)  
plt.ylabel('Tax (in USD)', fontsize=14)  

plt.grid(True, which='both', linestyle='--', linewidth=0.5)  

plt.show()
```

***Nhận xét biểu đồ #2 (Price vs Tax)***:

Biểu đồ **Price vs Tax** không cho thấy một mối quan hệ rõ ràng và dễ thấy giữa giá xe (Price) và mức thuế (Tax). Các nhận xét chính như sau:

1. **Phân cụm rõ ràng**: Có thể quan sát thấy các điểm dữ liệu phân bố thành nhiều cụm khác nhau. Điều này cho thấy mức thuế không đồng nhất và có thể phụ thuộc vào nhiều yếu tố khác ngoài giá xe, chẳng hạn như loại xe, khu vực hoặc chính sách thuế khác nhau.

2. **Mức thuế không tỉ lệ thuận với giá xe**: Mặc dù giá xe dao động rất lớn (từ vài ngàn đến 60,000 USD), mức thuế không tăng đáng kể theo giá. Thay vào đó, mức thuế chủ yếu tập trung ở khoảng 0 đến 300 USD, với một vài ngoại lệ có thuế cao hơn.

3. **Ngoại lệ bất thường**: Một số điểm ngoại lệ, đặc biệt là những điểm với mức thuế trên 400 USD, có thể đại diện cho các trường hợp đặc biệt như xe có hiệu suất cao hoặc xe được đánh thuế cao vì lý do cụ thể (ví dụ: xe sang, xe có kích thước lớn).

4. **Mức thuế thấp ở các xe giá cao**: Có nhiều điểm dữ liệu ở các xe có giá cao nhưng lại bị đánh thuế tương đối thấp (dưới 200 USD), cho thấy thuế không phải là yếu tố quyết định lớn đến giá xe.

**Kết luận**: **Mối quan hệ giữa giá xe và thuế không rõ ràng trong biểu đồ này**. Có thể cần thêm thông tin hoặc các yếu tố khác để giải thích tại sao mức thuế của xe lại phân bố thành các nhóm khác nhau và không tương xứng với giá xe.

```python
#3 Price vs MPG
plt.figure(figsize=(10, 6))  
plt.scatter(df['price'], df['mpg'], c='orange', edgecolors='black', alpha=0.7, s=60)  

plt.title('Scatter plot of Price vs MPG', fontsize=16, fontweight='bold') 
plt.xlabel('Price (in USD)', fontsize=14)  
plt.ylabel('MPG (Miles per Gallon)', fontsize=14)  
plt.grid(True, which='both', linestyle='--', linewidth=0.6)  

plt.show()

```

Biểu đồ thể hiện mối quan hệ giữa giá xe (**Price**) và số dặm đi được trên một gallon xăng (**MPG - Miles per Gallon**):

1. **Mối quan hệ không rõ ràng**: Biểu đồ cho thấy không có một mối quan hệ rõ ràng giữa giá xe và MPG. Những chiếc xe có giá thấp và trung bình có thể có MPG rất cao, trung bình, hoặc thấp. Điều này cho thấy yếu tố MPG không phải là yếu tố quyết định trực tiếp đến giá cả của xe. Hoặc có thể nói ảnh hướng không đáng kể.

2. **Tập trung nhiều xe ở mức MPG thấp đến trung bình**: Phần lớn các xe có MPG nằm trong khoảng từ 40 đến 60. Đây có thể là mức tiêu thụ nhiên liệu trung bình cho các xe bình thường trên thị trường, thể hiện sự phân bổ rộng rãi của các loại xe với mức tiêu hao nhiên liệu này.

3. **Xe đắt tiền thường có MPG trung bình đến thấp**: Khi giá xe tăng lên, MPG có xu hướng giảm dần. Điều này có thể giải thích bằng việc các xe có giá cao thường là các mẫu xe sang trọng, lớn hơn, hoặc có động cơ mạnh mẽ hơn, và do đó tiêu hao nhiên liệu nhiều hơn.

4. **Ngoại lệ MPG rất cao**: Có một vài điểm ngoại lệ với các giá trị MPG rất cao (> 100). Đây có thể là các xe hybrid hoặc xe điện, có khả năng tiết kiệm nhiên liệu rất cao, nhưng giá cả không nằm trong mức quá cao, cho thấy đây có thể là các mẫu xe tiết kiệm nhiên liệu nhưng không phải là các dòng xe sang trọng.

 Kết luận: 
Biểu đồ nhấn mạnh rằng **MPG** không phải là yếu tố quyết định chính về giá của xe. Mặc dù có sự phân bổ lớn của các xe có MPG trung bình, nhưng những chiếc xe có giá cao lại không nhất thiết phải có mức MPG cao. Các yếu tố khác như thương hiệu, dòng xe, tính năng công nghệ có thể đóng vai trò quan trọng hơn trong việc định giá.

```python

import plotly.express as px
import pandas as pd

# Đọc dữ liệu từ file
url = "https://raw.githubusercontent.com/thieu1995/csv-files/main/data/seaborn/penguins.csv"
df = pd.read_csv(url)

# Loại bỏ các giá trị null
df_cleaned = df.dropna()

# Sử dụng Plotly Express
fig = px.scatter_3d(df_cleaned, 
                    x='bill_length_mm', 
                    y='bill_depth_mm', 
                    z='flipper_length_mm',
                    color='body_mass_g', # Màu sắc dựa vào trọng lượng cơ thể
                    size='body_mass_g',  # Kích thước điểm dựa vào trọng lượng cơ thể
                    title='4D Scatter Plot of Penguin Measurements',
                    labels={
                        'bill_length_mm': 'Bill Length (mm)',
                        'bill_depth_mm': 'Bill Depth (mm)',
                        'flipper_length_mm': 'Flipper Length (mm)',
                        'body_mass_g': 'Body Mass (g)'
                    })
fig.write_html("penguin_4d_plot.html")



```
