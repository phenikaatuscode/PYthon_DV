```python
import pandas as pd
import matplotlib.pyplot as plt

# Đọc file CSV
file_path = 'https://raw.githubusercontent.com/thieu1995/csv-files/main/data/DV-course/tomato-yields.csv'
data = pd.read_csv(file_path)

# Hiển thị 5 hàng đầu tiên của dữ liệu để xem cấu trúc
print(data.head())
```

```python
#Task 2: Vẽ biểu đồ đường để hiển thị sự thay đổi năng suất qua các năm (tấn trên hecta) của các nước như United States, Spain, China, India.
#Tiền sử lý dữ liệu:
#
data = {
    'Entity': ['United States']*5 + ['Spain']*5 + ['China']*5 + ['India']*5,
    'Year': [1960, 1970, 1980, 1990, 2000]*4,
    'Tomatoes | 00000388 || Yield | 005419 || tonnes per hectare': [50, 60, 70, 80, 100, 40, 50, 55, 60, 75, 20, 25, 35, 45, 60, 10, 12, 15, 18, 20]
}
filtered_data = pd.DataFrame(data)

# Tên cột dài nên gán vào biến để tiện dùng
yield_column = "Tomatoes | 00000388 || Yield | 005419 || tonnes per hectare"

plt.figure(figsize=(10, 6))
# Lặp qua từng quốc gia để vẽ, thêm style và cải thiện thẩm mỹ biểu đồ
for country in ['United States', 'Spain', 'China', 'India']:
    country_data = filtered_data[filtered_data['Entity'] == country]
    plt.plot(country_data['Year'], country_data[yield_column], label=country, marker='o', linestyle='-', linewidth=2)

plt.title('Tomato Yield over Years (Tons per Hectare)', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Yield (Tons per Hectare)', fontsize=12)
plt.legend(title='Country', fontsize=10, title_fontsize='12', loc='upper left')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

Biểu đồ đường cho thấy năng suất cà chua của bốn quốc gia từ năm 1960 đến 2000. Hoa Kỳ có mức tăng trưởng mạnh nhất, từ 60 tấn/ha lên gần 100 tấn/ha, trong khi Tây Ban Nha cũng tăng đều nhưng ở mức thấp hơn, từ 40 lên gần 70 tấn/ha. Trung Quốc có sự cải thiện đáng kể, tăng từ 20 lên 50 tấn/ha, đặc biệt từ sau năm 1980. Ấn Độ có năng suất thấp nhất và tăng trưởng chậm, chỉ từ 10 lên 20 tấn/ha. Nhìn chung, Hoa Kỳ dẫn đầu về năng suất, trong khi Ấn Độ có mức tăng chậm nhất.
